```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("TGM 2020-2023_eng.csv", sep=";", decimal=",")

# Pilih kolom Reading Interest
col = "Tingkat Kegemaran Membaca (Reading Interest)"

# Histogram sebelum outlier removal
plt.figure(figsize=(8,4))
plt.hist(df[col], bins=20, edgecolor="black")
plt.title(f"Histogram {col} (Before Outlier Removal)")
plt.xlabel("Reading Interest")
plt.ylabel("Frequency")
plt.show()

# Hitung IQR
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data tanpa outlier
df_no_outlier = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Histogram sesudah outlier removal
plt.figure(figsize=(8,4))
plt.hist(df_no_outlier[col], bins=20, edgecolor="black", color="green")
plt.title(f"Histogram {col} (After Outlier Removal with IQR)")
plt.xlabel("Reading Interest")
plt.ylabel("Frequency")
plt.show()

print("Jumlah data awal:", len(df))
print("Jumlah data setelah outlier dihapus:", len(df_no_outlier))
print("Jumlah data outlier yang dihapus:", len(df) - len(df_no_outlier))
print("Batas bawah:", lower_bound)
print("Batas atas:", upper_bound)
```
